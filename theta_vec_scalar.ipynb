```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
sinv = np.loadtxt('theta/sin/vector', usecols=(3, 9, 10))
sins = np.loadtxt('theta/sin/scalar', usecols=(3, 9, 10))
```

```python
n = sinv[:, 0].size

data = sinv
for i in range(0, n, 8):
    plt.loglog(data[i:i+8, 1], data[i:i+8, 2], label='lx1=%d'%(data[i,0]+1))
plt.title('zsin.usr; vector version')
plt.ylabel('DOFS/s')
plt.xlabel('Time (s)')
plt.legend(frameon=False, loc=0)
```

```python
n = sins[:, 0].size

data = sins
for i in range(0, n, 8):
    plt.loglog(data[i:i+8, 1], data[i:i+8, 2], label='lx1=%d'%(data[i,0]+1))
plt.title('zsin.usr; scalar version')
plt.ylabel('DOFS/s')
plt.xlabel('Time (s)')
plt.legend(frameon=False, loc=0)
```
