```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
sinv = np.loadtxt('sin/vector', usecols=(3, 4, 10))
sins = np.loadtxt('sin/scalar', usecols=(3, 4, 10))
```

```python
n = sinv[:, 0].size

data = sinv
for i in range(0, n, 8):
    plt.semilogy([i for i in range(1, 9)], data[i:i+8, 2], label='lx1=%d'%(data[i,0]+1))
plt.title('zsin.usr; vector version')
plt.ylabel('DOFS/second')
plt.legend(frameon=False, loc=0)
```

```python
n = sins[:, 0].size

data = sins
for i in range(0, n, 8):
    plt.semilogy([i for i in range(1, 9)], data[i:i+8, 2], label='lx1=%d'%(data[i,0]+1))
plt.title('zsin.usr; scalar version')
plt.ylabel('DOFS/second')
plt.legend(frameon=False, loc=0)
```
